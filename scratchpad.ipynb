```python
sessions = list()
skips = list()
vocab = set()
session_ids = list()
cols = ['session_position', 'track_id_clean', 'skip_level'] #+ ['skip_1', 'skip_2', 'skip_3']
#skip_level = lambda df_: [4 - x for x in range(1, 4) if df_[f'skip_{x}'].bool() else 0] #0 - 3 in order of most severe skip
for s, df_s in tqdm(df.groupby(['session_id'])):
    df_s.sort_values(by = ['session_position'], axis=0, inplace=True)
    df_s['skip_level'] = ((df_s['skip_1'].astype(int) + df_s['skip_2'].astype(int) + df_s['skip_3'].astype(int)))
    df_s = df_s[cols]
    sessions.append(df_s['track_id_clean'].to_list())
    skips.append(df_s['skip_level'])
    vocab = vocab.union(set(sessions[-1]))
    session_ids.append(s)
print(len(session_ids) == len(sessions))
```

```python
import torch 
PADDING_ITEM = 0
mask_token = 2
def _mask_tracks(X, prop=0.15):
        Y = X[X != PADDING_ITEM]
        indices = torch.randperm(torch.numel(Y))
        indices = indices[:int(0.15//(1/indices.shape[0]))]
        Y[indices] = mask_token

        X[X != PADDING_ITEM] = Y

        return X

def _mask_tracks2(X, prop=0.15):
        X = X.detach().clone()
        dist = torch.ones(X.shape)
        dist[X == PADDING_ITEM] = 0
        #indices = torch.clamp(torch.multinomial(dist, int(X.shape[1] // (1/prop))), min=torch.ones((X.shape[0], 1)), 
                              #max=(torch.argmin(X, dim=1) - 1).unsqueeze(dim=-1)).long()
        indices = torch.multinomial(dist, 2)
        
        batch_idx = torch.tile(torch.linspace(0, X.shape[0] - 1, steps = X.shape[0]), (indices.shape[1],1)).long().T

        X[batch_idx, indices] = mask_token

        last_item = torch.argmin(X, dim=1)
        X[range(X.shape[0]), last_item] = mask_token

        return X
```

```python
y = torch.cat((torch.randint(300000, (64, 15)), torch.zeros((64, 6))), axis = 1)
z = torch.cat((torch.randint(300000, (64, 15)), torch.zeros((64, 6))), axis = 1)
m = _mask_tracks2(y)
#print(m)
#print(torch.argwhere(m == 2))
masked_targets = y[m == mask_token].view(64, 3)
print(masked_targets.shape)
last_item = torch.argmin(y, dim=1) - 1
#print(y[range(y.shape[0]), last_item])
masked_targets[:, -1] = z[range(z.shape[0]), last_item]

print(m)
print(masked_targets)
```
